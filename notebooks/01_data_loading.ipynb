# Data Loading and Validation


## Objective

This notebook validates the local dataset structure and confirms that the main competition files can be loaded correctly.

The goals are to:

- Verify the project paths
- Load a sample weekly training file
- Load `test.csv` and `test_input.csv`
- Confirm schema consistency
- Save a small metadata snapshot for traceability

```python
from pathlib import Path
from datetime import datetime
import json

import pandas as pd
import numpy as np
```

## Define Project Paths

We begin by defining the main project paths used throughout the notebook.

```python
PROJECT_ROOT = Path("..").resolve()
DATA_DIR = PROJECT_ROOT / "data" / "raw"

TRAIN_DIR = DATA_DIR / "train"
TEST_PATH = DATA_DIR / "test.csv"
TEST_INPUT_PATH = DATA_DIR / "test_input.csv"

REPORTS_DIR = PROJECT_ROOT / "reports"
SAMPLE_DIR = PROJECT_ROOT / "data" / "sample"

REPORTS_DIR.mkdir(exist_ok=True, parents=True)
SAMPLE_DIR.mkdir(exist_ok=True, parents=True)

print("PROJECT_ROOT:", PROJECT_ROOT)
print("TRAIN_DIR exists:", TRAIN_DIR.exists())
print("TEST_PATH exists:", TEST_PATH.exists())
print("TEST_INPUT_PATH exists:", TEST_INPUT_PATH.exists())
```

## List Training Files

We inspect the weekly training files available in the `train/` folder.

```python
train_files = sorted(TRAIN_DIR.glob("input_2023_w*.csv"))

print("Number of training files:", len(train_files))
print("First training files:")
for f in train_files[:10]:
    print("-", f.name)
```

##  Load a Sample Training File

To validate the raw schema, we load a single weekly training file.

```python
sample_train = pd.read_csv(train_files[0])

print("Sample train file:", train_files[0].name)
print("Sample train shape:", sample_train.shape)
print("\nSample train columns:")
print(sample_train.columns.tolist())

display(sample_train.head())
```

## Load Test Files

We load the provided competition test files:

- `test.csv`
- `test_input.csv`

```python
test = pd.read_csv(TEST_PATH)
test_input = pd.read_csv(TEST_INPUT_PATH)

print("Test shape:", test.shape)
print("Test input shape:", test_input.shape)

print("\nTest columns:")
print(test.columns.tolist())

print("\nTest input columns:")
print(test_input.columns.tolist())

display(test.head())
display(test_input.head())
```

## Schema Consistency Check

We compare the sample training file against `test_input.csv` to confirm whether they share the same input schema.

```python
train_cols = set(sample_train.columns)
test_input_cols = set(test_input.columns)

print("Columns only in train:", sorted(train_cols - test_input_cols))
print("Columns only in test_input:", sorted(test_input_cols - train_cols))
print("Same schema?", train_cols == test_input_cols)
```

##  Missing Value Snapshot

We inspect the top missing-value counts in the sample training data and test input.

```python
def top_missing(df, n=15):
    miss = df.isnull().sum()
    miss = miss[miss > 0].sort_values(ascending=False).head(n)
    return miss

print("Top missing values in sample_train:")
display(top_missing(sample_train))

print("Top missing values in test_input:")
display(top_missing(test_input))
```

## Target Indicator Check

The `player_to_predict` column appears central to the prediction task, so we inspect its distribution in the sample training file.

```python
if "player_to_predict" in sample_train.columns:
    print("player_to_predict value counts:")
    print(sample_train["player_to_predict"].value_counts(dropna=False))

    display(sample_train[sample_train["player_to_predict"] == True].head())
```

## Save Metadata Snapshot

To improve traceability, we save a small schema and dataset snapshot to the `reports/` directory.

```python
schema = {
    "created_at": datetime.now().isoformat(),
    "n_train_files": len(train_files),
    "sample_train_file": train_files[0].name if train_files else None,
    "sample_train_shape": list(sample_train.shape),
    "test_shape": list(test.shape),
    "test_input_shape": list(test_input.shape),
    "sample_train_columns": sample_train.columns.tolist(),
    "test_columns": test.columns.tolist(),
    "test_input_columns": test_input.columns.tolist(),
    "sample_train_dtypes": {c: str(t) for c, t in sample_train.dtypes.items()},
    "test_input_dtypes": {c: str(t) for c, t in test_input.dtypes.items()},
}

schema_path = REPORTS_DIR / "data_schema.json"
schema_path.write_text(json.dumps(schema, indent=2))

print("Saved schema snapshot to:", schema_path)
```

## Save Small Sample Files

We save small sample parquet files for faster experimentation in later notebooks.

```python
sample_train.head(5000).to_parquet(SAMPLE_DIR / "train_head_5k.parquet", index=False)
test_input.head(5000).to_parquet(SAMPLE_DIR / "test_input_head_5k.parquet", index=False)
test.head(5000).to_parquet(SAMPLE_DIR / "test_head_5k.parquet", index=False)

print("Saved sample files to:", SAMPLE_DIR)
```

In this notebook, we:

- validated the local competition file structure,
- loaded a sample weekly training file,
- loaded the provided test files,
- confirmed schema consistency,
- inspected missing values and the target indicator,
- and saved a metadata snapshot for traceability.

This notebook establishes the data-loading foundation for the rest of the project.
